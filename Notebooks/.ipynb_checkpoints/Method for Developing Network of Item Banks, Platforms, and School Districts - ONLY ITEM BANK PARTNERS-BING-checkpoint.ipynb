## Create Search Terms

For every ItemBankPartners and PlatformsPartners, combine with the school district to create search_terms for every record. For example, the search_terms for the first observation will be the following: <p>

```python
item_bank_partners = item_partners[['ItemBankPartners']]
item_bank_partners = item_bank_partners.rename(columns = {'ItemBankPartners': 'vendor'})
eligible_vendors = item_bank_partners.copy()
eligible_vendors  = eligible_vendors.drop_duplicates()
```

```python
eligible_vendors
```

```python
search_terms = []
for school in schools['School District']:
    for ven in eligible_vendors['vendor']: 
        term = "{} AND {} AND assessment".format(ven, school)
        search_terms.append(term)
search_terms
small = search_terms[:1]
```

```python
small
```

```python
returned_urls = []
phrase = []
title = []
for item in small:
    print(item)
    search_service = PyMsCognitiveWebSearch("a185959d275247529ba4bb965f9f56ce", item)
    first_result = search_service.search(limit=1, format='json')  # 1-50
    returned_urls.append(first_result[0].url)
    phrase.append(item)
    title.append(first_result[0].title)
    first_result = None
    search_service = None
returned_urls
phrase
title
```
